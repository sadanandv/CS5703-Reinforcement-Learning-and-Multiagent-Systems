
# Analysis of Football Dataset



## Importing Libraries

```python

import pandas as pd
```


## Loading the Dataset



```python

football_file_path = r"C:\Users\snucse\Desktop\OneDrive_1_12-14-2023\Datasets\football.csv"
football_df = pd.read_csv(football_file_path)
football_df.head()
```


## Number of Teams

Let's find out the number of teams that participated in the football matches.


```python

num_teams = football_df['Team'].nunique()
num_teams
```


## Sort Teams by Red Cards

I will sort the teams by the number of red cards they received.


```python

sorted_teams_red_cards = football_df[['Team', 'Red Cards']].sort_values(by='Red Cards', ascending=False).reset_index(drop=True)
sorted_teams_red_cards
```


## Mean Red Cards

Now, I will calculate the mean of red cards received by all the teams.


```python

mean_red_cards = football_df['Red Cards'].mean()
mean_red_cards
```


## Minimum and Maximum Number of Goals

Let's find the minimum and maximum number of goals scored by the teams.


```python

min_goals = football_df['Goals'].min()
max_goals = football_df['Goals'].max()
min_goals, max_goals
```


## Mean Shooting Accuracy

I will calculate the mean shooting accuracy based on the number of goals scored.


```python

mean_shooting_accuracy = (football_df['Shooting Accuracy'] * football_df['Goals']).sum() / football_df['Goals'].sum()
mean_shooting_accuracy
```


## Total Number of Shots

Finally, let’s calculate the total number of shots based on the number of players used.


```python

total_shots = (football_df['Total shots (inc. Blocked)'] / football_df['Players Used']).sum()
total_shots
```


# Analysis of Samp.User Dataset



## Importing Libraries

```python

import pandas as pd
```


## Loading the Dataset



```python

samp_user_file_path = r"C:\Users\snucse\Desktop\OneDrive_1_12-14-2023\Datasets\samp.user"
samp_user_df = pd.read_table(samp_user_file_path, sep='|', index_col='user_id')
samp_user_df.head()
```


## Min and Max Age per Occupation


```python

min_age_per_occupation = samp_user_df.groupby('occupation')['age'].min()
max_age_per_occupation = samp_user_df.groupby('occupation')['age'].max()
min_age_per_occupation, max_age_per_occupation
```


## Mean Age based on Occupation and Gender



```python

mean_age_by_occupation_gender = samp_user_df.groupby(['occupation', 'gender'])['age'].mean()
mean_age_by_occupation_gender
```
